```python
def ram_convertor(arg):
    return np.round((arg/1000),2)
```

```python
df['ram'] = df['ram'].apply(ram_convertor)
df.head()
```

```python
def battery_convertor(arg):
    return np.round((arg/1000),1)*1000
```

```python
df['battery_power'] = df['battery_power'].apply(battery_convertor)
df.rename(columns={'battery_power':'battery mAh'},inplace=True)
df.head()
```
